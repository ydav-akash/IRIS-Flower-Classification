```python
from sklearn.preprocessing import LabelEncoder
# Making Non Numerical Values Numerical iris-setosa = 0, iris-versicolor = 1, iris-verginica = 2
le = LabelEncoder()
Y = le.fit_transform(Y)
Y # Printing Transformed Y
```

```python
scaler = StandardScaler() # Initializing Scaler and Normal Variable
normal = Normalizer()
```

```python
data_scaler = scaler.fit_transform(X)
X = data_scaler
X.var() # Printing Variance in X Again
```

```python
# Splitting the Data into Training and Testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 1)
```

# **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression # Importing Logistic Regression Library
cl = LogisticRegression()
```

```python
cl.fit(X_train, Y_train) # Training the Model
```

```python
pred_test = cl.predict(X_test) # Testing the Model
pred_test
```

```python
# Creating the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm_test = confusion_matrix(Y_test, pred_test)
cm_test
```

```python
# Testing the Accuracy of the Model
test_accuracy = np.trace(cm_test)/np.sum(cm_test)
test_accuracy
```

```python
# Properly Formatting The Confusion Matrix
ax = plt.subplot()
sns.heatmap(cm_test, annot = True, fmt = 'g', ax = ax);
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(['iris_setosa', 'iris_versicolor', 'iris_verginica'])
ax.yaxis.set_ticklabels(['iris_setosa', 'iris_versicolor', 'iris_verginica'])
plt.show() # Displaying Confusion Matrix
```

# **Applying KNN Algorithm**

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import confusion_matrix
pred_prob = [] # Store all the Classifires Prediction Probablity in pred_prob Matrix
pred_prob.append(cl.predict_proba(X_test)) # Store Logistic Regression
plt.figure(figsize = (10, 12))
test_accuracy = []
for i in range(2, 15): # KNN for k = 2 to k = 14
  classifier = KNN(n_neighbors = i, metric = 'minkowski', p = 2)
  classifier.fit(X_train, Y_train)
  y_pred = classifier.predict(X_test) # Test the Model for each k and Make Confusion Matrix
  cm = confusion_matrix(Y_test, y_pred)
  print(cm)
  plt.subplot(2, 7, i)
  sns.heatmap(cm, annot = True, fmt = 'g', cmap = 'Blues', xticklabels = ['iris_setosa', 'iris-versicolor', 'iris-verginica'], yticklabels = ['iris_setosa', 'iris-versicolor', 'iris-verginica'])
  plt.title('Confusion Matrix')
  plt.show()
  test_acc = np.trace(cm_test)/np.sum(cm_test)
  test_accuracy.append(test_acc)
  pred_prob.append(classifier.predict_proba(X_test)) # Store the Classifier for the Specefic K in pred_prob Matrix
for i in range(1, 13):
  print(f"The accuracy for KNN model for k = {format(i + 1)} is {test_accuracy[i - 1]}")
```
